### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
#In the survey data, the DatabaseWantToWorkWith column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.
print(df.columns)

# Replace 'DatabaseDesireNextYear' with the actual column name if it's different
# Split the multiple selections (assuming comma-separated), count frequencies
desired_dbs = df['DatabaseWantToWorkWith'].dropna().str.split(';')
flat_list = [db.strip() for sublist in desired_dbs for db in sublist]
db_counts = pd.Series(flat_list).value_counts()

# Take the top 5 databases
top_5 = db_counts.head(5)

# Plot the pie chart
plt.figure(figsize=(6, 4))
plt.pie(top_5, labels=top_5.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Want to Work With')
plt.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle.
plt.show()


```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
# Process the DevType column
dev_types = df['DevType'].dropna().str.split(';')
flat_list = [role.strip() for sublist in dev_types for role in sublist]
dev_counts = pd.Series(flat_list).value_counts()

# Get the top 5 developer roles
top_5_dev_roles = dev_counts.head(5)

# Plot the pie chart
plt.figure(figsize=(6, 4))
plt.pie(top_5_dev_roles, labels=top_5_dev_roles.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Developer Roles Among Respondents')
plt.axis('equal')  # Keeps the pie chart round
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
##Write your code here
# Process the 'OpSysProfessional use' column
opsys = df['OpSysProfessional use'].dropna().str.split(';')
flat_opsys_list = [os.strip() for sublist in opsys for os in sublist]
opsys_counts = pd.Series(flat_opsys_list).value_counts()

# Get the top operating systems (adjust top N if needed)
top_opsys = opsys_counts.head(5)

# Plot the pie chart
plt.figure(figsize=(6, 4))
plt.pie(top_opsys, labels=top_opsys.index, autopct='%1.1f%%', startangle=140)
plt.title('Top Operating Systems Used Professionally')
plt.axis('equal')  # Keeps the pie chart round
plt.show()
```

### Task 2: Additional Visualizations and Comparisons


##### 2.1 Pie Chart for Top 5 Programming Languages Respondents Have Worked With


The `LanguageHaveWorkedWith` column contains the programming languages that respondents have experience with. We’ll plot a pie chart to display the composition of the top 5 languages.



```python
##Write your code here
# Process the 'LanguageHaveWorkedWith' column
languages = df['LanguageHaveWorkedWith'].dropna().str.split(';')
flat_lang_list = [lang.strip() for sublist in languages for lang in sublist]
language_counts = pd.Series(flat_lang_list).value_counts()

# Get the top 5 languages
top_languages = language_counts.head(5)

# Plot the pie chart
plt.figure(figsize=(6, 4))
plt.pie(top_languages, labels=top_languages.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Programming Languages Respondents Have Worked With')
plt.axis('equal')  # Ensures the pie chart is a circle
plt.show()
```

##### 2.2 Pie Chart for Top Collaboration Tools used in Professional Use


Using the `NEWCollabToolsHaveWorkedWith` column, we’ll identify and visualize the top collaboration tools respondents use in their professional work.



```python
##Write your code here
# Process the 'NEWCollabToolsHaveWorkedWith' column
collab_tools = df['NEWCollabToolsHaveWorkedWith'].dropna().str.split(';')
flat_tools_list = [tool.strip() for sublist in collab_tools for tool in sublist]
collab_counts = pd.Series(flat_tools_list).value_counts()

# Get the top 5 collaboration tools
top_collab_tools = collab_counts.head(5)

# Plot the pie chart
plt.figure(figsize=(6, 4))
plt.pie(top_collab_tools, labels=top_collab_tools.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Collaboration Tools Used by Respondents')
plt.axis('equal')  # Keeps the pie chart round
plt.show()
```
